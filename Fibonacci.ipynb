##FIBONACCI

```python
import time
import matplotlib.pyplot as plt
#n = int(input())
start_time = time.time()
n = 35
fibo = []
t = []
x,y,sum = 0,1,0
for i in range(n):
	x = y
	y = sum
	sum = x + y
	fibo.append(y)
	t.append(time.time() - start_time)
print (fibo)
print (t)
plt.ylabel("Tiempo")
plt.xlabel("Fibonacci")
plt.plot(fibo,t)
```
